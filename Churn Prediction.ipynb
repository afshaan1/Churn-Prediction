## 4) Data Preparation

### 4.1) Identifying Target Variable and splitting the data into Train-Test

```python
df.head()
```

```python
y = df['Churn']
X = df[['gender','SeniorCitizen','Partner','Dependents','tenure','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaperlessBilling','PaymentMethod','MonthlyCharges','TotalCharges']]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, train_size=0.75, random_state=40)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

### 4.2) Seperating Categorical and Numerical Columns

```python
X_train.head()
```

```python
X_train.dtypes
```

```python
X_train_cat = X_train.select_dtypes(include=['object'])
X_train_cat.head()
```

```python
X_train_num = X_train.select_dtypes(include=['int64','float64'])
X_train_num.head()
```

### 4.3) Scaling the numerical features

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_num_rescaled = pd.DataFrame(scaler.fit_transform(X_train_num), columns=X_train_num.columns, index=X_train_num.index)

X_train_num_rescaled.head()
```

```python
X_train_num.describe()
```

```python
print("Number of Numerical Features:", scaler.n_features_in_)
print("Mean of each column:", scaler.mean_)
print("Std of each column:", np.sqrt(scaler.var_))
```

### 4.4) Apply OneHotEncoding on Categorical Columns

```python
X_train_cat.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)

X_train_cat_ohe = pd.DataFrame(encoder.fit_transform(X_train_cat),
                               columns=encoder.get_feature_names_out(X_train_cat.columns),
                               index = X_train_cat.index)

X_train_cat_ohe.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first', sparse_output=False)

X_train_cat_ohe = pd.DataFrame(encoder.fit_transform(X_train_cat),
                               columns=encoder.get_feature_names_out(X_train_cat.columns),
                               index = X_train_cat.index)

X_train_cat_ohe.head()
```

```python
encoder.categories_
```

### 4.5) Concatenating the Encoded Categorical Features and Rescaled Numerical Features

```python
X_train_transform = pd.concat([X_train_num_rescaled, X_train_cat_ohe], axis=1)
X_train_transform.head()
```

```python
X_train_transform
```

### 4.6) Test Data

```python
X_test.info()
```

```python
X_test.head()
```

```python
X_test_cat = X_test.select_dtypes(include=['object'])
X_test_cat.head()
```

```python
X_test_cat
```

```python
X_test_num = X_test.select_dtypes(include=['int64','float64'])
X_test_num.head()
```

```python
X_test_num_rescaled = pd.DataFrame(scaler.transform(X_test_num), columns = X_test_num.columns, index = X_test_num.index)
X_test_num_rescaled.head()
```

```python
X_test_cat_ohe = pd.DataFrame(encoder.fit_transform(X_test_cat),
                              columns = encoder.get_feature_names_out(X_test_cat.columns),
                              index = X_test_cat.index)
X_test_cat_ohe.head()
```

```python
X_test_transform = pd.concat([X_test_num_rescaled,X_test_cat_ohe], axis=1)
X_test_transform.head()
```

```python
X_test_transform
```

### 5) Model Building

```python
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train_transform, y_train)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train_transform, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train_transform, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(max_iter=150)
classifier.fit(X_train_transform, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train_transform, y_train)
```

### 6) Prediction and Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
y_test_pred = knn_classifier.predict(X_test_transform)
knn_accurac = accuracy_score(y_test,y_test_pred)
print(knn_accurac)
```

```python
y_test_pred = svc.predict(X_test_transform)
svc_accurac = accuracy_score(y_test,y_test_pred)
print(svc_accurac)
```

```python
y_test_pred = dt_classifier.predict(X_test_transform)
dt_accurac = accuracy_score(y_test,y_test_pred)
print(dt_accurac)
```

```python
y_test_pred = classifier.predict(X_test_transform)
log_accurac = accuracy_score(y_test,y_test_pred)
print(log_accurac)
```

```python
y_test_pred = rfc.predict(X_test_transform)
random_forest_accurac = accuracy_score(y_test,y_test_pred)
print(random_forest_accurac)
```

```python
print(classification_report(y_test, y_test_pred))
```
